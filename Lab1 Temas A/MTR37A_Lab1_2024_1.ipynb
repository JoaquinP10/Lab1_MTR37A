Escribir una función en python llamada `simular_movimiento` que recibe una lista de comandos (por ejemplo: ['avanzar', 'giro_izquierda', 'avanzar']) e imprime la posición final de un robot en coordenadas X,Y

ejemplo:

````python
#-----------------------------
from math import sin, cos, radians ## importar funciones de la libreria math
ang = 0 ## orientación inicial

def simular_movimiento(cmds):
  global ang, x, y ## ang, x, y son variables globales (fuera de la función)
  for acc in cmds: #lee a través de la lista
    if acc=='avanza':
      [dx, dy] = [-round(sin(radians(ang))), round(cos(radians(ang)))] ##declara los desplazamientos en función al ángulo
                                                                       ## más información en README.md
      x = x + dx ##actualiza la posición sumando desplazamiento
      y = y + dy
    elif acc=='retrocede':
      [dx, dy] = [round(sin(radians(ang))), -round(cos(radians(ang)))]
      x = x + dx
      y = y + dy
    elif acc=='giro_derecha': ##para giros, actualiza la orientación
      ang = (ang + 270) % 360
    elif acc=='giro_izquierda':
      ang = (ang + 90) % 360
    else:
      print("Comando incorrecto") ## en caso no se ingrese un comando válido
  return f'x,y = ({x},{y})' ##retorna una cadena con los valores impresos


#-----------------------------
x,y = (0,0)
cmds = ['avanza']
print(simular_movimiento(cmds))  # Retorna x,y = (0,1)
cmds = ['giro_derecha', 'avanza']
print(simular_movimiento(cmds))  # Retorna x,y = (1,1)
cmds = ['giro_izquierda', 'retrocede']
print(simular_movimiento(cmds))  # Retorna x,y = (1,0)
````

##Lectura de sensores NMEA

Los sensores GPS son altamente empleados en robótica móvil debido a su capacidad de determinar la ubicación (proceso de localización) del vehículo en un entorno. Estos sensores proporcionan la información mediante cadenas de texto en un formato conocido como NMEA (National Marine Electronics Association), el cual se muestra a continuación:

```
...
```

Cada dato se estuctura de la siguiente manera:


```
$PGPSP,d1,d2,d3,lf1,lf2,lf3,f,f,f,f,f,f,f,f,f,f*xx

donde
d1 : Milisegundos desde la mañana del domingo en GMT
d2 : Número de semanas desde el 1 de enero de 1980 en GMT
d3 : Estado del GPS: [0x000000xx] número de satélites utilizados, [0x0000xx00] tipo de corrección, [0x00xx0000] indicadores de estado

lf1: WGS84 Latitude en grados
lf2: WGS84 Longitude en grados

f: [valores adicionales retornados por el sensor]

xx: Valor Checksum en hexadecimal
```

El valor del CRC (checksum: suma de comprobación de paquete) es un número entero de 8 bits y se obtiene calculando el **OR exclusivo** de todos los **bytes** intermedios y sin incluir los caracteres $ y *

en python la operación or exclusivo se peude realizar con el operador ^

```
xor = A ^ B
```

1. Desarrollar una función `calcular_CRC` que reciba una cadena de texto en formato NMEA y calcule el CRC de la trama. Pueden emplear la página https://nmeachecksum.eqth.net/ para validar el CRC optenido

TIP 1:
Para convertir una cadena de texto en una secuencia de **bytes** pueden emplear el metodo .encode() de la clase cadena

```
str = 'hola mundo'

>>> print(str) # Imprime una cadena
'hola mundo'

>>> print(str.encode()) # Imprime una secuencia de bytes
b'hola mundo'

```



````python
#-----------------------------
data = "$PGPSP,250472200,2140,197396,-12.06866870,-77.07954170,93.68,84.76,1.32,1.57,3.26,-0.03,0.19,0.06,0.46,29.0,249603.8609,18*5d" ## ejemplo

def calcular_CRC(nmea):
  ## Declara CRC. Se empieza con 0 ya que p^0 = p
  crc = 0x00
  ## Extrae la cadena entre $ y *
  start = data.find("$")
  end = data.find("*")
  ndata = data[start+1:end]

  ndata_bytes = ndata.encode() ##obtiene la secuencia de bytes de la cadena
  for byte in ndata_bytes:
    crc = crc ^ byte ## se opera entre todos los bytes dentro de la secuencia
  hex_crc = hex(crc) ##convierte crc a su representación en hexadecimal (la salida es una cadena)

  return hex_crc[2:] ## finalmente, se devuelve la cadena (sin el 0x)

print(calcular_CRC(data))

#-----------------------------
````
